### 1) Creation of entangled states using Haddamard gate


```python
def target_state1(n, indices_of_marked_elements):
    
    # Create a quantum circuit on 2 qubits
    qc = QuantumCircuit(2, 2)
    
    # step 1: apply Hadamard gates on first gate qubits
    qc.h(0)
    
    # step 2: cnot gate on the second gate
    qc.cx(0,1)
    qc.x(1)
        
    # step 3: measure all qubits
    qc.measure(range(2), range(2))
  
    return qc

mycircuit = target_state1(6, [1, 42])
mycircuit.draw(output='text')
```

```python
from qiskit import Aer, execute
simulator = Aer.get_backend('qasm_simulator')
counts = execute(mycircuit, backend=simulator, shots=1000).result().get_counts(mycircuit)
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

The above graph shows that the resulting sate is a superposition of states |01> and |10> where |01> and |10> occur with the same probability. We can look at the probability distribution using different number of simulation shots as follows:

```python
from qiskit import Aer, execute
simulator = Aer.get_backend('qasm_simulator')
counts = execute(mycircuit, backend=simulator, shots=1).result().get_counts(mycircuit)
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```
